```python
"""
10. Множество всех купленных некоторым клиентом товаров за 2 года назовём потребительской
корзиной для этого клиента.
Укажите пару товаров, которая встречается в наибольшем числе потребительских корзин
"""
from itertools import combinations
from collections import Counter

# Группируем уникальные товары по клиентам
baskets = df.groupby('client_id')['product_name'].apply(set)

pairs_counter = Counter()
for basket in baskets:
    # Создаем все возможные пары из товаров в корзине
    if len(basket) >= 2:
        pairs = combinations(sorted(basket), 2)
        pairs_counter.update(pairs)

pairs_counter.most_common(1)
```

```python
"""
11. Множество всех купленных некоторым клиентом товаров за 2 года назовём потребительской
корзиной для этого клиента.
Рассмотрим клиента с “client_id = 1011”. Укажите “client_id” клиента, у которого
потребительская корзина имеет наибольшее пересечение с потребительской корзиной клиента с
“client_id = 1011”.

12. Множество всех купленных некоторым клиентом товаров за 2 года назовём потребительской
корзиной для этого клиента.
Рассмотрим клиента с “client_id = 1116”. Укажите “client_id” клиента, у которого
потребительская корзина имеет наибольшее пересечение с потребительской корзиной клиента с
“client_id = 1116”.
"""
def find_most_similar(target_id, baskets_dict):
    target_basket = baskets_dict[target_id]
    max_intersect = -1
    similar_client = None
    
    for client_id, basket in baskets_dict.items():
        if client_id == target_id:
            continue
        
        # Считаем размер пересечения множеств
        intersect_size = len(target_basket.intersection(basket))
        
        if intersect_size > max_intersect:
            max_intersect = intersect_size
            similar_client = client_id
        # Если пересечения равны, обычно берут ID по возрастанию, 
        # но в олимпиадах ответ часто уникален.
            
    return similar_client

# Превращаем серию в словарь для скорости
baskets_dict = baskets.to_dict()

# 11. Для клиента 1011
ans_11 = find_most_similar(1011, baskets_dict)

# 12. Для клиента 1116
ans_12 = find_most_similar(1116, baskets_dict)

ans_11, ans_12
```
